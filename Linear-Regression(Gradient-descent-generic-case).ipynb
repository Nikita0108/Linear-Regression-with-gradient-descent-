```python
def step_gradient(X,Y,learning_rate,m):
    m_slope=np.zeros(14)
    s=0
    M=379
    x=0
    y=0
    new_m=np.zeros(14)
    for i in range(M):
        y=Y[i]
        s=0
        for k in range(14):
            s=s+m[k]*X[i][k]
        for j in range(14):
            x=X[i,j]
            m_slope[j]+=(-2/M)*(y-s)*x
            
    for l in range(14):  
        new_m[l]=m[l]-learning_rate*m_slope[l]
    
    return new_m
```

```python
def gd(X,Y,no_iterations,learning_rate):
    m=np.zeros(14)
    for i in range(no_iterations):
        m=step_gradient(X,Y,learning_rate,m)
        print(i,"Cost: ",cost(X,Y,m))
    return m
```
